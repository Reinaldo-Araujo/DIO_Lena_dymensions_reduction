```python
import numpy as np
import urllib.request
from PIL import Image
import matplotlib.pyplot as plt

# Baixa a imagem Lena
url = "https://raw.githubusercontent.com/opencv/opencv/master/samples/data/lena.jpg"
urllib.request.urlretrieve(url, "lena.jpg")

# Carrega a imagem como um array numpy
img = Image.open("lena.jpg")
img_array = np.array(img)

# Converte a imagem para níveis de cinza utilizando coeficientes de intensidade de luninância
gray_img = np.dot(img_array[..., :3], [0.2989, 0.5870, 0.1140])

# Redução de dimensionalidade (escalando para níveis de 0 a 255)
gray_img = (gray_img / np.max(gray_img)) * 255
gray_img = gray_img.astype(np.uint8)

# Converte a imagem para preto e branco
bw_img = (gray_img > 128) * 255
bw_img = bw_img.astype(np.uint8)

# Salva as imagens
Image.fromarray(gray_img).save("lena_gray.jpg")
Image.fromarray(bw_img).save("lena_bw.jpg")

# Visualiza as imagens
plt.figure()
plt.imshow(img_array)
plt.title("Imagem Original")
plt.figure()
plt.imshow(gray_img, cmap='gray')
plt.title("Imagem Cinza")
plt.figure()
plt.imshow(bw_img, cmap='gray')
plt.title("Imagem Preto e Branco")
plt.show()
```
